# Create a 3d plot 

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

print("x:")
print(x.shape)
print(x[:5])

print("y:")
print(y.shape)
print(y[:5])
```

```python
X,Y = np.meshgrid(x,y)

print("X:")
print(X.shape)
print(X[:5, :5])

print("Y:")
print(Y.shape)
print(Y[:5, :5])
```

```python
Z = X**2 + Y**3

print("Z:")
print(Z.shape)
print(Z[:5, :5])
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='3d')
ax.plot_surface(X,Y,Z,cmap=plt.cm.jet)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
