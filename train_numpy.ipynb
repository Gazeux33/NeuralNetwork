```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Create a 1D array of numbers from 0 to 9
x1 = np.arange(10)
print(x1)
```

```python
# 2. Create a 3×3 numpy array
x2 = np.arange(9).reshape(3,3)
print(x2)
```

```python
# 3. Create a 3×3 numpy of 0
x3 = np.zeros((3,3))
print(x3)
```

```python
# 4. Create a 3×3 numpy array of all True’s
x4 = np.full((3,3), True, dtype=bool)
print(x4)


x5 = np.full((3,3), 99, dtype=int)
print(x5)

x6 = np.ones((3,3))
print(x6)
```

```python
x7 = np.array([[1,2,3],[1,2,3],[1,2,3]], dtype=np.float32)
print(x7)
```

```python
X = np.random.rand(10,2)
print("X:")
print(X[:5])

Y = np.random.choice(np.arange(4), size=(X.shape[0],1))
print("Y:")
print(Y[:5])
```

```python
plt.scatter(X[:,0], X[:,1],c=Y, cmap='summer')
```
